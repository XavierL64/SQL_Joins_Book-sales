```python
# Select all data in order_history with an order_id found with the query above.
SELECT * 
FROM order_history 
WHERE order_id = 4412;
```

```python
#  Count number of returned orders
SELECT COUNT(*) 
FROM order_history
WHERE status_id = 6
```

```python
# Join the order_history and cust_order tables for a specific order_id
SELECT * 
FROM cust_order
JOIN order_history ON order_history.order_id = cust_order.order_id
WHERE cust_order.order_id = 4412;
```

```python
# Join the order_history and cust_order tables, aggregating the data per customer and sorting by number of returned orders 
SELECT cust_order.customer_id, COUNT(order_history.order_id) AS returned_orders  
FROM cust_order
JOIN order_history ON order_history.order_id = cust_order.order_id
WHERE order_history.status_id = 6
GROUP BY cust_order.customer_id
ORDER BY returned_orders DESC; 
```

```python
# Filter for customers that have returned 2 or more orders and add contact details for each customer by building on previous query and joining data with the customer table
SELECT customer.first_name, customer.last_name, customer.email, cust_order.customer_id, COUNT(order_history.order_id) AS returned_orders  
FROM cust_order
JOIN order_history ON order_history.order_id = cust_order.order_id
LEFT JOIN customer ON customer.customer_id = cust_order.customer_id -- using left join to return users that have returned items but for which there is no customer information
WHERE order_history.status_id = 6
GROUP BY cust_order.customer_id, customer.first_name, customer.last_name, customer.email
HAVING COUNT(order_history.order_id) > 1
ORDER BY returned_orders DESC; 
```
